```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('Teams.csv')
```

```python
df_2018 = df[df['yearID'] == 2018] #2018 baseball team stats
win_total = df_2018['W']
plt.hist(win_total, bins=5, color='blue')
plt.xlabel('Win Total', fontsize=16)
plt.ylabel('Number Of Teams', fontsize=16)
plt.title('2018 MLB team win totals', fontsize=16, color='blue')
plt.show()
```

```python
homeruns = df_2018['HR']
plt.scatter(x=homeruns, y=win_total, color='red')
plt.ylim([0,120])
plt.xlim([100,260])
plt.ylabel('Win Total', fontsize=16)
plt.xlabel('HR total', fontsize=16)
plt.title('Homeruns vs. Win Total', color = 'red', fontsize=20)
plt.show()
```

```python
plt.boxplot(win_total)
plt.ylabel('2018 MLB Team Win Totals', fontsize=14, color='black')
plt.show()
```

```python
era = df_2018['ERA']

plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.scatter(x=homeruns, y=win_total, color='red')
plt.ylim([0,120])
plt.xlim([100,260])
plt.ylabel('Win Total', fontsize=16)
plt.xlabel('Team HR total', fontsize=16)
plt.title('Homeruns vs. Win Total', color = 'red', fontsize=20)

plt.subplot(1,2,2)
plt.scatter(x=era, y=win_total, color='blue')
plt.ylim([0,120])
plt.ylabel('Win Total', fontsize=16)
plt.xlabel('Team ERA', fontsize=16)
plt.title('ERA vs. Win Total', color = 'blue', fontsize=20)

plt.show()

```
